```python
'''목표: valset의 값을 기반으로, 각 방정식의 계산값을 구하고 그 결과를 .csv 파일에 저장'''

# Read data using Pandas
import pandas as pd

data =  pd.read_csv('../1. 초기데이터/Data_Set_LDL-C_valset.csv')
```

```python
'''Extract data'''
import numpy as np

#> Extract Train Data
TC = data.iloc[:,[3]]
TC = np.array(TC)

HDL_C = data.iloc[:,[4]]
HDL_C = np.array(HDL_C)

TG = data.iloc[:,[6]]
TG = np.array(TG)

Non_HDL_C = data.iloc[:,[7]]
Non_HDL_C = np.array(Non_HDL_C)

#> Extract Target data
LDL_C = data.iloc[:,[5]]
LDL_C = np.array(LDL_C)
```

```python
'''Calculate Formula'''

#> Fridewald Formula
LDL_CF = TC - HDL_C - (TG/5)



######################################################################################### 
#> Marin Formula

marin_csv =  pd.read_csv('../1. 초기데이터/Marin Forumla Table.csv' , header = None )  ## marin_Formual_table Load
marin_data = marin_csv.to_numpy() ## DataFrame -> numpy.array
 
#>> Non_HDL_C 를 formula_table 에 맞춰 인덱스 계산 = Non_HDL_C_idx
Non_HDL_C_idx = np.where(Non_HDL_C< 100, 0 , Non_HDL_C )
Non_HDL_C_idx = np.where(np.logical_and( Non_HDL_C_idx >= 100 , Non_HDL_C_idx < 130 ), 1 , Non_HDL_C_idx )
Non_HDL_C_idx = np.where(np.logical_and( Non_HDL_C_idx >= 130 , Non_HDL_C_idx < 160 ), 2 , Non_HDL_C_idx )
Non_HDL_C_idx = np.where(np.logical_and( Non_HDL_C_idx >= 160 , Non_HDL_C_idx < 190 ), 3 , Non_HDL_C_idx )
Non_HDL_C_idx = np.where(np.logical_and( Non_HDL_C_idx >= 190 , Non_HDL_C_idx < 220 ), 4 , Non_HDL_C_idx )
Non_HDL_C_idx = np.where(Non_HDL_C_idx >= 220, 5 , Non_HDL_C_idx )

#>> TC 를 formula_table 에 맞춰 인덱스 계산 = TC_idx
TC_idx = np.where(np.logical_and( TC >= 7 , TC < 50 ), 0, TC )
TC_idx = np.where(np.logical_and( TC_idx >= 50 , TC_idx < 57 ), 1, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 57 , TC_idx < 62 ),2, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 62 , TC_idx < 67 ),3, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 67 , TC_idx < 72 ), 4,  TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 72 , TC_idx < 76 ),5, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 76 , TC_idx < 80 ),6, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 80 , TC_idx < 84 ),7, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 84 , TC_idx < 88 ),8, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 88 , TC_idx < 93 ),9, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 93, TC_idx < 97 ),10, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 97, TC_idx < 101 ),11, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 101, TC_idx < 106 ),12, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 106, TC_idx < 111 ),13, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 111, TC_idx < 116 ),14, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 116, TC_idx < 121 ),15, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 121, TC_idx < 127 ),16, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 127, TC_idx < 133 ),17, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 133, TC_idx < 139 ),18, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 139, TC_idx < 147 ),19, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 147, TC_idx < 155 ),20, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 155, TC_idx < 164 ),21, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 164, TC_idx < 174 ),22, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 174, TC_idx < 186 ),23, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 186, TC_idx < 202 ),24, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 202, TC_idx < 221 ),25, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 221, TC_idx < 248 ),26, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 248, TC_idx < 293 ),27, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 293, TC_idx < 400 ),28, TC_idx )
TC_idx = np.where(np.logical_and( TC_idx >= 400, TC_idx < 13975 ),29, TC_idx )


#>> float 형을 int형으로 바꿔줌 
Non_HDL_C_idx = Non_HDL_C_idx.astype(np.int64) 

#>> 다차원 배열을 1차원 배열로 변환
Non_HDL_C_idx = Non_HDL_C_idx.flatten()
TC_idx = TC_idx.flatten()

#>> Non_HDL_C 와 TG 값에 따른 val_data 의 AF _TABLE <- LDL-CM = Non-HDL-C − (TG/AFij)
af_table = []

for i in range(5483):
    af_table.append(marin_data[TC_idx[i]][Non_HDL_C_idx[i]]) 

af_table=[ [i] for i in af_table]
AF = np.array(af_table) # numpy 형식으로 바꿈

# 구현하기
LDL_CM = Non_HDL_C - (TG / AF)


#> Simpson Formula
temp1 = TC / 0.948
temp2 = HDL_C / 0.971
temp3 = (TG / 8.56) + ((TG * Non_HDL_C) / 2140) - (TG**2 / 16100)
LDL_CS = temp1 - temp2 - temp3 - 9.44
```

```python
'''Export estimate data of each formula'''
df_CF = pd.DataFrame(LDL_CF ,columns=['Fridewald_Formula'] )
df_CM = pd.DataFrame(LDL_CM ,columns=['Marin Formula'] )
df_CS = pd.DataFrame(LDL_CS ,columns=['Simpson Formula'])

df_result =  pd.concat([df_CF, df_CM, df_CS], axis = 1)
#df_result =  pd.concat([df_CF, df_CM, df_CS, df_DNN], axis = 1)
df_result.to_csv('../1. 초기데이터/Data_Set_LDL-C-formula.csv')
```
